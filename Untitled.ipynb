```python
import statsmodels.formula.api as smf
```

```python
lm = smf.ols(formula="sepal_length~sepal_width", data = data).fit()
```

```python
lm.params
```

El modelo lineal predictivo sera sepal_length = 7.032594 + 0.047537 * sepal_width

```python
lm.pvalues
```

```python
lm.rsquared
```

```python
lm.rsquared_adj
```

```python
lm.summary()
```

```python
sepal_length = lm.predict(pd.DataFrame(data["sepal_width"]))
sepal_length
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
data.plot(kind = "scatter", x = "sepal_width", y = "sepal_length")
plt.plot(pd.DataFrame(data["sepal_width"]), sepal_length, c="red", linewidth = 2)
```

```python
lm2 = smf.ols(formula="petal_length~petal_width", data = data).fit()
```

```python
lm2.params
```

```python
lm2.rsquared
```

```python
lm2.rsquared_adj
```

```python
petal_length = lm2.predict(pd.DataFrame(data["petal_width"]))
petal_length
```

```python
%matplotlib inline
data.plot(kind = "scatter", x = "petal_width", y = "petal_length")
plt.plot(pd.DataFrame(data["petal_width"]), petal_length, c="red", linewidth = 2)
```
